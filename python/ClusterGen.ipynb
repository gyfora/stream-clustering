```python
from scipy import random, linalg
import pandas as pd
```

```python
def generate_clusters(num_clusters,seed = 0, dist = 2, var = 1, size = 1000, fig_dir = "", means = None, cvars = None):
    clusters = [] 
    mean = (1,1)
    
    if seed != 0:
        np.random.seed(seed)
        
    #Generate the individual clusters
    for i in range(num_clusters):
        if cvars != None:
            cov = np.array([[cvars[i][0],0],[0,cvars[i][1]]])
        else:
            A = random.rand(2,2)*np.array([[var,0],[0,var]])
            cov = numpy.dot(A,A.transpose())
        if means != None:
            mean = means[i]
        else:
            mean = mean + dist*(2*random.rand(2)-(1,1))
        sample = np.random.multivariate_normal(mean,cov = cov, size = size)
        clusters.append(sample)
        
    #Merge the clusters, standardize and attach the labels    
    data = np.concatenate(clusters,0)
    data = (data-np.mean(data,axis=0))/np.var(data,axis=0)
    data = np.hstack((data,np.array(sorted(size*list(range(1,num_clusters+1)))).reshape(-1,1)))
    
#     #Generate noise
#     if noise > 0:
#         noise_size = num_clusters*size*noise
#         x_scale = max(data[:,0])-min(data[:,0])
#         noise_x = np.random.random(noise_size)*x_scale - x_scale/2
#         y_scale = max(data[:,1])-min(data[:,1])
#         noise_y = np.random.random(noise_size)*y_scale - y_scale/2
#         noise = np.vstack((noise_x,noise_y, np.repeat(-1,num_clusters*size*noise))).T
#         data = np.vstack((data,noise))
    
    data_df = pd.DataFrame(data,columns=["x","y","cluster"])
    
    plot_data(data_df)
  
    if fig_dir != "":
        plt.savefig(fig_dir)
        
    return data_df

def create_stream(data,split_at=-1, mix = 0.5):
    mix = 1-mix
    if split_at==-1:
        return data.ix[np.random.permutation(data.index)]
    else:
        before = data[data.cluster < split_at]
        before = before.ix[np.random.permutation(before.index)]
        after = pd.concat((before[int(before.shape[0]*mix):],data[data.cluster >= split_at]))
        before = before[:int(before.shape[0]*mix)]
        return pd.concat((before.ix[np.random.permutation(before.index)],after.ix[np.random.permutation(after.index)]))
        

def plot_data(data_df):
    p = plt.scatter(data_df.x,data_df.y,c=data_df.cluster, cmap=mpl.cm.coolwarm)
    p.figure.set_size_inches(8,8)
    
def plot_progression(data, nsplit):
    f, axes = plt.subplots(1, nsplit)
    
    xlim = (min(data.x),max(data.x))
    ylim = (min(data.y),max(data.y))
    
    points = list(map(lambda x: int(x), np.linspace(0,data.shape[0],nsplit+1)))
    
    for (i,ax) in enumerate(axes):
        d = data[points[i]:points[i+1]]
        ax.scatter(d.x,d.y,c=d.cluster, cmap=mpl.cm.coolwarm)
        plt.xlim(xlim)
        plt.ylim(ylim)
    
    f.set_size_inches(nsplit*6.5,6)
        
def evaluate(data, prediction, fig_dir = ""):    
    # Two subplots, unpack the axes array immediately
    f, (ax1, ax2) = plt.subplots(1, 2)
    ax1.scatter(data.x,data.y,c=data.cluster, cmap=mpl.cm.coolwarm)
    ax1.set_title('Original')
    ax2.scatter(prediction.x,prediction.y,c=prediction.cluster, cmap=mpl.cm.coolwarm)
    ax2.set_title('Prediction')
    
    f.set_size_inches(13,6)
    
#     noise_index = data[data.cluster == -1].index
    
    if fig_dir != "":
        plt.savefig(fig_dir)
    
    print("Original SSE: " + str(compute_sse(data)))
    print("Prediction SSE: " + str(compute_sse(prediction)))

def compute_sse(data):
#     data = data.drop(noise_index)
    centers = data.groupby("cluster").sum()/data.groupby("cluster").count()
    data_c = data.join(centers, on="cluster", rsuffix="_c")
    return sum(np.sqrt(np.square(data_c["x_c"]-data_c["x"])+np.square(data_c["y_c"]-data_c["y"])))
```

```python
data = generate_clusters(4,dist = 10,var = 1.5,seed=6, size=100)
```
